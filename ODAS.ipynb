```python
import yfinance as yf
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
import matplotlib as plt
import requests
from bs4 import BeautifulSoup as bs
```

```python
ODAS = yf.download('ODAS.IS', start = '2022-01-01', end = '2022-12-31')
ODAS
```

```python
#satir silme 
ODAS = ODAS [ODAS.index > '2022-01-01']
ODAS
```

```python
#sutun silme 
ODAS = ODAS[['Adj Close']]
ODAS
```

```python
#isim degistirme
ODAS.columns = ['Kapanis TRY']
ODAS
```

```python
norm = ODAS.iloc[0, 0]
ODAS['Normalize'] = ODAS['Kapanis TRY']/norm
ODAS
```

```python
ODAS['Normalize'].plot()
```

```python
figure1 = px.line(ODAS, y = ("Normalize"))
figure2 = px.bar(ODAS, y = ("Normalize"))
figure3 = px.scatter(ODAS, y = ("Normalize"))
figure1.show()
figure2.show()
figure3.show()
```

```python
fig_1 = px.line(ODAS, y = ("Normalize"), width = 1000, height = 500, title = ('whatever is meant to be written'), 
        hover_data = {"Normalize" : ':.2f', "Kapanis TRY" : ':.2f'},
        labels = {'Date' : 'Tarih', "Normalize" : 'Bist TRY'})
fig_1.show()
```

```python
ODAS ['degisim'] = ODAS['Normalize'].diff(1)
```

```python
ODAS["d_yon"] = ODAS['degisim'].map (lambda x : 'dusen gun' if x < 0 else 'yukselen gun') 
ODAS
```

```python
ODAS['degisim%'] = ODAS['degisim']/ODAS["Normalize"].shift(1)*100
ODAS
```

```python
px.bar(ODAS, y = "degisim",
       color = "d_yon", color_discrete_map = {'dusen gun' : 'orange', "yukselen gun" : 'navy'})
```

```python
px.histogram(ODAS['degisim%'], nbins = 10)
```

```python
Kur = yf.download('TRY=X', start = '2022-01-01', end = '2022-12-31')
Kur = Kur[['Adj Close']]
Kur.columns = ['USD/TRY']
Kur['Normalize USD/TRY'] = Kur['USD/TRY']/Kur.iloc[0, 0]    
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(x = ODASK.index, y = ODASK['Normalize'], name = "Normalize ODAS"))
fig.add_trace(go.Scatter(x = ODASK.index, y = ODASK['Normalize USD/TRY'], name = "Normalize USD/TRY"))
fig.update_layout(title = 'Normalize ODAS ve Normalize Kur',
                  xaxis_title = 'Tarih', xaxis_dtick = 'M1',
                  yaxis_title = "Kapanis (TRY)")
```

```python
indirme_listesi = ['ODAS.IS', 'TRY=X']
veri_2022 = pd.DataFrame()

for i in indirme_listesi:
  data = yf.download(i, start = '2022-01-01', end = '2022-12-31')
  data = data['Adj Close'].rename(i)
  veri_2022 = pd.concat([veri_2022, data], axis = 1)  

veri_2022
```

```python
veri_2022.columns = ['Kapanis (TRY)', 'USD/TRY']
veri_2022 = veri_2022[veri_2022['Kapanis (TRY)']>0]
veri_2022
```

For loop 

```python
indirme_listesi = ['ODAS.IS', 'TRY=X']
veri_2023 = pd.DataFrame()

for i in indirme_listesi:
  data = yf.download(i, start = '2023-01-01', end = '2023-04-30')
  data = data['Adj Close'].rename(i)
  veri_2023 = pd.concat([veri_2023, data], axis = 1)  

veri_2023
```

```python
veri_2023.columns = ['Kapanis (TRY)', 'USD/TRY']
veri_2023 = veri_2023[veri_2023['Kapanis (TRY)']>0]
veri_2023
```

```python
veri = pd.concat([veri_2022, veri_2023], axis = 0)
veri
```

```python
veri['Normalize'] = veri['Kapanis (TRY)']/veri.iloc[0, 0]
#veri['Normalize USD/TRY'] = veri['USD/TRY']/veri.iloc[0, 0]
veri['Normalize USD/TRY'] = veri['USD/TRY']/veri.loc[veri.index[0], 'USD/TRY']
veri['degisim'] = veri['Normalize'].diff(1)
veri['d_yon'] = veri['degisim'].map(lambda x: 'dusen gun' if x < 0 else 'yukselen gun')
veri['degisim%'] = (veri['degisim']/veri['Normalize'].shift(1))*100
veri
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(x = veri.index, y = veri['Normalize'], name = "Normalize ODAS"))
fig.add_trace(go.Scatter(x = veri.index, y = veri['Normalize USD/TRY'], name = "Normalize USD/TRY"))
fig.update_layout(title = 'Normalize ODAS ve Normalize Kur',
                  xaxis_title = 'Tarih', xaxis_dtick = 'M1',
                  yaxis_title = "Kapanis (TRY)")
```

```python
veri['ay_yil'] = veri.index.strftime('%Y-%m')
veri
```

```python
veri_as = veri.pivot_table(values = ['Kapanis (TRY)', 'USD/TRY'], index = 'ay_yil', aggfunc = 'last')
veri_as
```

```python
veri_as['degisim'] = veri_as['Kapanis (TRY)'].diff(1)
veri_as['d_yon'] = veri_as['degisim'].map(lambda x: 'dusen gun' if x < 0 else 'yukselen gun')
veri_as['degisim%'] = (veri_as['degisim']/veri_as['Kapanis (TRY)'].shift(1))
veri_as
```

```python
px.bar(veri_as, title = 'ODAS', y = "degisim%",
       color = "d_yon", color_discrete_map = {'dusen gun' : 'red', "yukselen gun" : 'green'},
       text_auto = ".0%")
```
