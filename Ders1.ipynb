```python
import yfinance as yf
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
import matplotlib as plt
import requests
from bs4 import BeautifulSoup as bs
```

```python
D100 = yf.download('XU100.IS', start = '2022-01-01', end = '2022-12-31')
D100
```

```python
D100.head(10)
```

```python
D100.info()
```

```python
D100.describe()
```

```python
#satir silme 
D100 = D100 [D100.index > '2022-01-01']
D100
```

```python
#sutun silme 
D100 = D100[['Adj Close']]
D100
```

```python
#isim degistirme
D100.columns = ['Kapanis TRY']
D100
```

```python
norm = D100.iloc[0, 0]
D100['Normalize'] = D100['Kapanis TRY']/norm
D100
```

```python
D100['Normalize'].plot()
```

```python
figure1 = px.line(D100, y = ("Normalize"))
figure2 = px.bar(D100, y = ("Normalize"))
figure3 = px.scatter(D100, y = ("Normalize"))
figure1.show()
figure2.show()
figure3.show()
```

```python
fig_1 = px.line(D100, y = ("Normalize"), width = 1000, height = 500, title = ('whatever is meant to be written'), 
        hover_data = {"Normalize" : ':.2f', "Kapanis TRY" : ':.2f'},
        labels = {'Date' : 'Tarih', "Normalize" : 'Bist TRY'})
fig_1.show()
```

```python
D100 ['degisim'] = D100['Normalize'].diff(1)
```

```python
D100["d_yon"] = D100['degisim'].map (lambda x : 'dusen gun' if x < 0 else 'yukselen gun') 
D100
```

```python
D100['degisim%'] = D100['degisim']/D100["Normalize"].shift(1)*100
D100
```

```python
px.bar(D100, y = "degisim",
       color = "d_yon", color_discrete_map = {'dusen gun' : 'orange', "yukselen gun" : 'navy'})
```

```python
px.histogram(D100['degisim%'], nbins = 10)
```

```python
indirme_listesi = ['XU100.IS', 'TRY=X']
veri_2022 = pd.DataFrame()

for i in indirme_listesi:
  data = yf.download(i, start = '2022-01-01', end = '2022-12-31')
  data = data['Adj Close'].rename(i)
  veri_2022 = pd.concat([veri_2022, data], axis = 1)  

veri_2022
```

```python
veri_2022.columns = ['Kapanis (TRY)', 'USD/TRY']
veri_2022 = veri_2022[veri_2022['Kapanis (TRY)']>0]
veri_2022
```

For loop 

```python
indirme_listesi = ['XU100.IS', 'TRY=X']
veri_2023 = pd.DataFrame()

for i in indirme_listesi:
  data = yf.download(i, start = '2023-01-01', end = '2023-04-30')
  data = data['Adj Close'].rename(i)
  veri_2023 = pd.concat([veri_2023, data], axis = 1)  

veri_2023
```

```python
veri_2023.columns = ['Kapanis (TRY)', 'USD/TRY']
veri_2023 = veri_2023[veri_2023['Kapanis (TRY)']>0]
veri_2023
```

```python
veri = pd.concat([veri_2022, veri_2023], axis = 0)
veri
```

```python
veri['Normalize'] = veri['Kapanis (TRY)']/veri.iloc[0, 0]
#veri['Normalize USD/TRY'] = veri['USD/TRY']/veri.iloc[0, 0]
veri['Normalize USD/TRY'] = veri['USD/TRY']/veri.loc[veri.index[0], 'USD/TRY']
veri['degisim'] = veri['Normalize'].diff(1)
veri['d_yon'] = veri['degisim'].map(lambda x: 'dusen gun' if x < 0 else 'yuksek gun')
veri['degisim%'] = (veri['degisim']/veri['Normalize'].shift(1))*100
veri
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(x = veri.index, y = veri['Normalize'], name = "Normalize BIST100"))
fig.add_trace(go.Scatter(x = veri.index, y = veri['Normalize USD/TRY'], name = "Normalize USD/TRY"))
fig.update_layout(title = 'Normalize BIST100 ve Normalize Kur',
                  xaxis_title = 'Tarih', xaxis_dtick = 'M1',
                  yaxis_title = "Kapanis (TRY)")
```

```python
veri['ay_yil'] = veri.index.strftime('%Y-%m')
veri
```

```python
veri_as = veri.pivot_table(values = ['Kapanis (TRY)', 'USD/TRY'], index = 'ay_yil', aggfunc = 'last')
veri_as
```

```python
veri_as['degisim'] = veri_as['Kapanis (TRY)'].diff(1)
veri_as['d_yon'] = veri_as['degisim'].map(lambda x: 'dusen gun' if x < 0 else 'yukselen gun')
veri_as['degisim%'] = (veri_as['degisim']/veri_as['Kapanis (TRY)'].shift(1))
veri_as
```

```python
px.bar(veri_as, y = "degisim%",
       color = "d_yon", color_discrete_map = {'dusen gun' : 'red', "yukselen gun" : 'green'},
       text_auto = ".0%")
```

```python
veri_as2 = veri.resample('BM', convention = 'end').asfreq()
veri_as2
```

```python
url = 'https://www.tcmb.gov.tr/wps/wcm/connect/TR/TCMB+TR/Main+Menu/Istatistikler/Enflasyon+Verileri/Tuketici+Fiyatlari'
```

```python
response = requests.get(url)
response.text
```

```python
soup = bs(response.text, 'html.parser')
table = soup.find('table', {'summary' : 'Fiyat Endeksi (Tüketici Fiyatları)(2003=100)'})
table
```

```python
tufe = pd.read_html(str(table))
tufe = pd.DataFrame(tufe[0])
tufe = tufe[['Unnamed: 0', 'TÜFE (Aylık % Değişim)']]
tufe.columns = ['index', 'tufe%']
tufe['index'] = tufe['index'].str[-4:] +'-'+ tufe['index'].str[:2]
tufe.set_index('index', inplace = True)
tufe['tufe%'] = tufe['tufe%']/100
tufe = tufe[tufe.index.str[:4]>'2021']
tufe
```

```python
veri_enf = pd.concat([veri_as, tufe], axis = 1)
veri_enf 
```

```python
veri_enf['degisim%_enf'] = ((1 + veri_enf['degisim%'])/(1 + veri_enf['tufe%']))-1
veri_enf['d_yon_enf'] = veri_enf['degisim%_enf'].map(lambda x: 'dusen gun' if x < 0 else 'yukselen gun')
veri_enf
```

```python
px.bar(veri_enf, y = "degisim%_enf",
       color = "d_yon_enf", color_discrete_map = {'dusen gun' : 'red', "yukselen gun" : 'green'},
       text_auto = ".0%")
```
